## 여러 한글 파일을 하나의 한글 파일로 합치기

- 작업폴더 1, 작업폴더 2 안에 각각 '제1회', '제2회', .. 와 같이 폴더가 들어있고 그 안에 A, B, C 한글파일이 들어있는 상황
- 목표 : 각 작업폴더에서 A끼리만 모아서 역순으로 하나의 한글 파일로 합치기

[코드 출처]("https://one-click.tistory.com/10")

```python

import win32com.client as win32
import os
import fnmatch

# 작업할 전체 폴더
dir_original ="C:/project_sbh/의뢰" 
# 작업폴더 1(사례형)
dir_case ="C:/project_sbh/의뢰/1-12회 사례형(공법, 형사법, 민사법)"
# 작업폴더 2(선택형)
dir_select ="C:/project_sbh/의뢰/1-12회 선택형"
```

```python
dir = dir_select # 선택형

select_gong, select_minsa, select_hyungsa = [], [], []

for folder in os.listdir(dir):
    folder = dir+"/"+folder
    # filename=fnmatch.filter(os.listdir(dir),'*.hwp')
    for hwpfile in os.listdir(folder):
        # print(folder+"/"+hwpfile)
        if '공법' in hwpfile:
            select_gong.append(folder+"/"+hwpfile)
        elif '민사법' in hwpfile:
            select_minsa.append(folder+"/"+hwpfile)
        elif '형사법' in hwpfile:
            select_hyungsa.append(folder+"/"+hwpfile)

print(select_gong, select_minsa, select_hyungsa)


dir = dir_case # 사례형

case_gong, case_minsa, case_hyungsa = [], [], []

for folder in os.listdir(dir):
    folder = dir+"/"+folder
    # filename=fnmatch.filter(os.listdir(dir),'*.hwp')
    for hwpfile in os.listdir(folder):
        # print(folder+"/"+hwpfile)
        if '공법' in hwpfile:
            case_gong.append(folder+"/"+hwpfile)
        elif '민사법' in hwpfile:
            case_minsa.append(folder+"/"+hwpfile)
        elif '형사법' in hwpfile:
            case_hyungsa.append(folder+"/"+hwpfile)

print(case_gong, case_minsa, case_hyungsa)
```

```python
import shutil

def move_files_to_folder(file_list, destination_folder):
    os.makedirs(destination_folder, exist_ok=True)
    for file_path in file_list:
        file_name = os.path.basename(file_path)
        destination_path = os.path.join(destination_folder, file_name)
        shutil.move(file_path, destination_path)

# 폴더를 생성하고 파일 이동하는 코드. 묶을 파일끼리
move_files_to_folder(select_gong, "공법_선택형")
move_files_to_folder(select_minsa, "민사법_선택형")
move_files_to_folder(select_hyungsa, "형사법_선택형")

move_files_to_folder(case_gong, "공법_사례형")
move_files_to_folder(case_minsa, "민사법_사례형")
move_files_to_folder(case_hyungsa, "형사법_사례형")
```

```python
# 파일을 역순으로 읽어내기. re_sorted하지 않을 경우 10회, 11회, 12회, 1회, .. 와 같이 순서가 엉망이 됨. 
def re_sorted(data):
    def extract_number(s):
        return int(s.split('회')[0][1:])
    sorted_data = sorted(os.listdir(dir_original+"/"+folder), key=extract_number, reverse=True)  # 순서를 거꾸로 정렬
    return sorted_data

folder_all = ['공법_선택형', '민사법_선택형', '형사법_선택형','공법_사례형','민사법_사례형','형사법_사례형']

# 선택형은 파일 앞에 '파일명'이 쓰여진 페이지 추가(페이지에 몇회인지 안나와있기 때문)
for folder in folder_all[:3]:
    print(folder, '----STARTED')
    ###############
    hwp = win32.gencache.EnsureDispatch("HWPFrame.HwpObject")
    i = 0
    for file in re_sorted(os.listdir(dir_original+"/"+folder)):
        if i == 0:
            print(file)
            hwp.Open(dir_original+"/"+folder+"/"+file, "HWP", "forceopen:true")
            hwp.MovePos(2, 0, 0)
            ####
            hwp.HAction.GetDefault("InsertText", hwp.HParameterSet.HInsertText.HSet)
            hwp.HParameterSet.HInsertText.Text = file[:-4]+": 아래 페이지부터 시작합니다."
            hwp.HAction.Execute("InsertText", hwp.HParameterSet.HInsertText.HSet)
            hwp.Run("BreakPage")
            ####
            hwp.MovePos(3, 0, 0)


        else:
            print(file)
            ####
            hwp.Run("BreakPage")
            hwp.HAction.GetDefault("InsertText", hwp.HParameterSet.HInsertText.HSet)
            hwp.HParameterSet.HInsertText.Text = file[:-4]+": 아래 페이지부터 시작합니다."
            hwp.HAction.Execute("InsertText", hwp.HParameterSet.HInsertText.HSet)
            ####
            hwp.HAction.GetDefault("InsertFile", hwp.HParameterSet.HInsertFile.HSet)
            option = hwp.HParameterSet.HInsertFile
            option.filename = dir_original+"/"+folder+"/"+file
            option.KeepSection = 1
            option.KeepCharshape = 1
            option.KeepParashape = 1
            option.KeepStyle = 1
            hwp.HAction.Execute("InsertFile", hwp.HParameterSet.HInsertFile.HSet)
            hwp.MovePos(3, 0, 0)

        i = i + 1

    hwp.HAction.GetDefault("FileSaveAs_S", hwp.HParameterSet.HFileOpenSave.HSet)
    option = hwp.HParameterSet.HFileOpenSave
    option.Attributes = 0
    option.filename = dir_original+"/"+folder+".hwp"
    option.Format = "HWP"
    hwp.HAction.Execute("FileSaveAs_S", hwp.HParameterSet.HFileOpenSave.HSet)

    hwp.Quit()
    print(folder, '----FINISHED')



# 그대로 합치기(페이지에 몇회인지 나와있음)

for folder in folder_all[3:]:
    print(folder, '----STARTED')

    ###############
    hwp = win32.gencache.EnsureDispatch("HWPFrame.HwpObject")
    i = 0
    for file in re_sorted(os.listdir(dir_original+"/"+folder)):
        if i == 0:
            print(file)
            hwp.Open(dir_original+"/"+folder+"/"+file, "HWP", "forceopen:true")
            hwp.MovePos(3, 0, 0)


        else:
            print(file)
            hwp.HAction.GetDefault("InsertFile", hwp.HParameterSet.HInsertFile.HSet)
            option = hwp.HParameterSet.HInsertFile
            option.filename = dir_original+"/"+folder+"/"+file
            option.KeepSection = 1
            option.KeepCharshape = 1
            option.KeepParashape = 1
            option.KeepStyle = 1
            hwp.HAction.Execute("InsertFile", hwp.HParameterSet.HInsertFile.HSet)
            hwp.MovePos(3, 0, 0)

        i = i + 1

    hwp.HAction.GetDefault("FileSaveAs_S", hwp.HParameterSet.HFileOpenSave.HSet)
    option = hwp.HParameterSet.HFileOpenSave
    option.Attributes = 0
    option.filename = dir_original+"/"+folder+".hwp"
    option.Format = "HWP"
    hwp.HAction.Execute("FileSaveAs_S", hwp.HParameterSet.HFileOpenSave.HSet)

    hwp.Quit()
    print(folder, '----FINISHED')
```
